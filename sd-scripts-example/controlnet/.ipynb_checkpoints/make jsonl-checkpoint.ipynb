```python
import os
import json
from tqdm import tqdm

# 디렉토리 경로 설정
root_dir = '/data/sd-dataset/celeba_controlnet'
image_dir = 'images'
conditioning_dir = 'conditioning_images'
meta_file = 'metadata.jsonl'

# metadata.jsonl 파일 열기
with open(os.path.join(root_dir, meta_file), 'w') as outfile:
    # 이미지 디렉토리에서 파일 목록 가져오기
    for image_file in tqdm(os.listdir(os.path.join(root_dir, image_dir))[:100]):
        if image_file.endswith('.jpg'):
            # 파일 이름에서 확장자 제거하여 기본 이름 얻기
            basename = os.path.splitext(image_file)[0]
            
            # 이미지 파일 경로
            image_rel_path = os.path.join(image_dir, image_file)
            image_abs_path = os.path.join(root_dir, image_rel_path)
            
            # 캡션 파일 경로
            caption_rel_path = os.path.join(image_dir, basename + '.txt')
            caption_abs_path = os.path.join(root_dir, caption_rel_path)
            
            # 조건 파일 경로
            conditioning_rel_path = os.path.join(conditioning_dir, image_file)
            conditioning_abs_path = os.path.join(root_dir, conditioning_rel_path)
            
            # 캡션 파일 읽기
            with open(caption_abs_path, 'r') as caption_file:
                caption = caption_file.read().strip()
            
            # JSON 객체 생성
            metadata = {
                "conditioning_image": basename,
                "image": basename,
                "text": caption
            }
            
            # JSON 객체를 문자열로 변환 후 파일에 쓰기
            outfile.write(json.dumps(metadata) + '\n')
```

```python
import json

# JSONL 파일의 내용을 저장할 리스트
metadata_list = []

# JSONL 파일 읽기
with open(os.path.join(root_dir, meta_file), 'r') as infile:
    for line in infile:
        # 각 줄을 JSON 객체로 변환하여 리스트에 추가
        metadata_list.append(json.loads(line))

metadata_list
```
